```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
from config1 import gkey

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

output_data_file = "output_data/cities.csv"
```

```python
#RANGES
lat_range = (-90, 90)
lng_range = (-180, 180)

#CITIES AND LAT/LNG LIST
lat_lngs = []
cities = []


# CREATING RANDOM COORDENATES
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


#INDENTYFING NEARST CITY
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # ADDING TO THE LIST
    if city not in cities:
        cities.append(city)

        
len(cities)
#print(cities)
```

```python
# CREATING THE SET OF LIST TI STORED THE DATA

City_name = []
Temperature = [] 
Humidity = []
Cloudness = []
Wind_speed = []
Lat = []
Lng = []
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

query_url = f"{url}appid={gkey}&units={units}&q="

response = requests.get(query_url + "Alameda").json()

#print(f" is.. {json.dumps(response, indent=2)}") 
```

```python
sum= 0

for xcity in cities:    
    response = requests.get(query_url + xcity).json()

    try:  
        
        City_name.append(response["name"][0])
        Lat.append(response["coord"]["lat"][0])
        Lng.append(response["coord"]["lng"][0])    
        Temperature.append(response["weather"]["main"]["temp"][0])
        Humidity.append(response["weather"]["main"]["humidity"][0])
        Cloudness.append(response["clouds"]["all"][0])
        Wind_speed.append(response["wind"]["speed"][0])

        City_sum = response["name"]                  
        sum+= 1    
        print(f"Sequence: {sum} - checking: {City_sum} ")

    except: 
        print("City not founded")



```

```python
weather_df = pd.DataFrame({
    "City": City_name,
    "Lat" : Lat,
    "Lng" : Lng,
    "Temp": Temperature,
    "Humidity" : Humidity,
    "Cloudness": Cloudness,
    "Wind speed": Wind_speed   
 })

weather_df
```

```python
plt.scatter(weather_df["Lat"], weather_df["Temp"], marker="o", color="red")

plt.xlabel("Latitude")
plt.ylabel("Temperature in F")

plt.title("Temperature Vs Latitude")
plt.grid(True)


plt.savefig("TemperatureInWorldCities.png")
plt.show()
```
